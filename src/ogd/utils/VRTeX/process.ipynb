## Create activity graphs

```python
player_number = 2

def create_player_plot(player_number, overlay_data_list):
    
    # create a 4x5 grid of subplots

    fig = make_subplots(rows=1, cols=1, shared_yaxes=True, shared_xaxes=True, subplot_titles=('View Data', 'Left Hand', 'Right Hand'))
    
    fig.add_trace(go.Scatter(x=df_pos_rot[player_number]['timesincelaunch'][:-1], y=1-df_pos_rot[player_number]['dot_product'][:-1], mode='lines'), row=1, col=1)
    #fig.add_trace(go.Scatter(x=left_pos_pairwise[player_number]['timesincelaunch'][:-1], y=left_pos_pairwise[player_number]['dot_product'][:-1], mode='lines'), row=2, col=1)
    #fig.add_trace(go.Scatter(x=right_pos_pairwise[player_number]['timesincelaunch'][:-1], y=right_pos_pairwise[player_number]['dot_product'][:-1], mode='lines'), row=3, col=1)
    
    fig.update_layout(title=f'View Activity for Single Player', showlegend=False, yaxis_range=[0,1])
    
    colors = ["orange", "green", "red"]
    player_id = gaze_package_lst_splt[player_number]["player_id"].iloc[0]
    for i,event_data in enumerate(overlay_data_list):
        other_data_for_player = event_data[event_data['player_id'] == player_id]
        event_times_for_player = other_data_for_player['timesincelaunch'].tolist()
        print(f"There are {len(event_data)} events in group {i}")
        print(f"percent waddle happens overall: {len(event_times_for_player)/len(gaze_package_lst_splt[player_number]['player_id'])*100}")
        line_col = colors[i % len(overlay_data_list)]
        line_width = 1 + i
        print(f"line color is now {line_col}")
        if event_times_for_player is not []:
            for event_time in event_times_for_player:
                fig.add_vline(x=event_time, line_width=line_width, line_dash="dash", line_color=line_col)
        
    fig.show()

create_player_plot(player_number,[waddle_lst])
```
